```python
import requests

# URL de ton endpoint FastAPI
url = "http://127.0.0.1:8000/predict"

# Exemple avec des films spécifiques
payload = {"user_id": 1}

# Requête POST
response = requests.post(url, json=payload)

# Afficher la réponse JSON
print(response.json())
```

```python
import pandas as pd
response = requests.post(url, json=payload)
data = response.json()

# Transformer en DataFrame
top_reco_df = pd.DataFrame(data["recommendations"])
print(top_reco_df)
```

```python
top_reco_df = top_reco_df.sort_values(by="predicted_rating", ascending=False)[["title", "predicted_rating"]]
print(top_reco_df)
```
