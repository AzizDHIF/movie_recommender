```python
import requests

# URL de ton endpoint FastAPI
url = "http://127.0.0.1:8081/predict"

# Exemple avec des films spécifiques
payload = {"user_id": 1}

# Requête POST
response = requests.post(url, json=payload)

# Afficher la réponse JSON
print(response.json())
```

```python
import pandas as pd
response = requests.post(url, json=payload)
data = response.json()

# Transformer en DataFrame
top_reco_df = pd.DataFrame(data["top_recommendations"])
print(top_reco_df)
```

```python
top_reco_df = top_reco_df.sort_values(by="predicted_rating", ascending=False)[["title", "predicted_rating"]]
print(top_reco_df)
```

```python
import requests

url = "http://127.0.0.1:8081/predict"

# Exemple : utilisateur 1 qui a noté certains films
payload = {
    "user_id": 1,
    "movie_ids": [2329.0, 923.0]  # films que l'utilisateur a vus / notés
}

response = requests.post(url, json=payload)

# Vérifier le status
print("STATUS CODE:", response.status_code)

# Afficher la réponse JSON si OK
if response.status_code == 200:
    print(response.json())
else:
    print("Erreur :", response.text)
```

```python
# Obtenir top-N après mise à jour
payload_topn = {"user_id": 1}  # juste l'utilisateur, pas de films spécifiques
response_topn = requests.post(url, json=payload_topn)
print(response_topn.json())
```

```python
# Transformer en DataFrame
import pandas as pd
data = response_topn.json()

# Transformer en DataFrame
top_reco_df = pd.DataFrame(data["top_recommendations"])
print(top_reco_df)
```
