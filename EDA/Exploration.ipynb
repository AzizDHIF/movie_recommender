```python
from google.cloud import bigquery

project_id="students-group2"
client = bigquery.Client(project=project_id)


query = """
SELECT *
FROM `master-ai-cloud.MoviePlatform.ratings`
LIMIT 10
"""
df_ratings = client.query(query).to_dataframe()
```
